```python
import numpy as np
from emukit.core import ContinuousParameter, ParameterSpace

np.random.seed(seed_value)

# space = ParameterSpace([ContinuousParameter('x1', 2000, 50000),
#                         ContinuousParameter('x2', 400, 16000),
#                         ContinuousParameter('x3', 0, 0.6),
#                         #ContinuousParameter('x4', 10, 400),
#                         #ContinuousParameter('x5', 300000, 6000000),
#                         #ContinuousParameter('x6', 30, 600)])
#                         ContinuousParameter('x7', 0, 110000),
#                         ContinuousParameter('x8', 0, 400),
#                         ContinuousParameter('x9', 15, 60)])

space = ParameterSpace([ContinuousParameter(feat, stats[feat].loc['min'], stats[feat].loc['max']) for feat in stats.columns])
```

```python
Y = df['GDT_TS'].to_numpy().reshape(-1,1)
Y.shape
```

```python
#X = df.drop('GDT_TS', axis=1).to_numpy()
X = arr_scaled
X.shape
```

```python
from GPy.models import GPRegression
from emukit.model_wrappers import GPyModelWrapper

model_gpy = GPRegression(X,Y)
model_emukit = GPyModelWrapper(model_gpy)
model_emukit.optimize()
```

```python
from emukit.sensitivity.monte_carlo import MonteCarloSensitivity

senstivity = MonteCarloSensitivity(model = model_emukit, input_domain = space)

main_effects, total_effects, _ = senstivity.compute_effects(num_monte_carlo_points = 10000)
#main_effects_minmaxscaler, total_effects_minmaxscaler, _ = senstivity.compute_effects(num_monte_carlo_points = 10000)
#main_effects_stdscaler, total_effects_stdscaler, _ = senstivity.compute_effects(num_monte_carlo_points = 10000)
```

```python
print(main_effects)
print(total_effects)
```

```python
main_effects_list = []
total_effects_list = []
```

```python
main_effects_list.append(main_effects)
total_effects_list.append(total_effects)
```

```python
import numpy as np

def get_feat_values(d):
    feat_vals = []
    for feat, val in d.items():
        feat_vals.append(val[0])

    return feat_vals

def get_avg_effects(main_effects_list, total_effects_list):
    # Main effects
    data_main_effects = []
    for main_effects in main_effects_list:
        data_main_effects.append(get_feat_values(main_effects))

    data_main_effects = np.array(data_main_effects)
    print(data_main_effects)

    feats = [feat for feat, val in main_effects_list[0].items()]

    main_effects_avg = {}
    for i in range(len(feats)):
        main_effects_avg[feats[i]] = np.array([data_main_effects[:, i].mean()])

    # Total effects
    data_total_effects = []
    for total_effects in total_effects_list:
        data_total_effects.append(get_feat_values(total_effects))

    data_total_effects = np.array(data_total_effects)
    print(data_total_effects)

    feats = [feat for feat, val in total_effects_list[0].items()]

    total_effects_avg = {}
    for i in range(len(feats)):
        total_effects_avg[feats[i]] = np.array([data_total_effects[:, i].mean()])

    return main_effects_avg, total_effects_avg
```

```python
import matplotlib.pyplot as plt

def grouped_plot(data, title):
    tmp = {}
    feats = []
    for method, d in data.items():
        feats = d.keys()
        tmp[method] = [d[feat][0] for feat in feats]

    x = np.arange(len(feats))  # the label locations
    width = 0.17  # the width of the bars
    multiplier = 0

    fig, ax = plt.subplots(layout='constrained')

    for method, feat_values in tmp.items():
        offset = width * multiplier
        rects = ax.bar(x + offset, feat_values, width, label=method)
        #ax.bar_label(rects, padding=6)
        multiplier += 1

    ax.set_ylabel('Effects Value')
    ax.set_title(title)
    #ax.set_xticks(x + width, feats) # use if multiple methods
    ax.set_xticks(x, feats)
    ax.legend()#loc='upper left')
    #ax.set_ylim(0, 250)

    plt.show()
```

```python
d = {'Monte Carlo': main_effects}
grouped_plot(d, "Main effects")

d = {'Monte Carlo': total_effects}
grouped_plot(d, "Total effects")
```

```python
main_effects_avg, total_effects_avg = get_avg_effects(main_effects_list, total_effects_list)

d = {'Monte Carlo': main_effects_avg}
grouped_plot(d, "Main effects")

d = {'Monte Carlo': total_effects_avg}
grouped_plot(d, "Total effects")
```
